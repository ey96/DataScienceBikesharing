## Convert start positions of trip data to points

```python
# returns "Point(longtitude, latitude)" of the starting position of a rental
def make_point(row):
    return Point(row.longitude_start, row.latitude_start)


```

```python
# Go through every row, and make a point out of its lat and lon
trips_june["geometry"] = trips_june.apply(make_point, axis=1)

# It doesn't come with a CRS because it's a CSV, so it has to be set
trips_june.crs = {'init': 'epsg:4326'}
```

```python
# drop unnecessary attributes for the scope of this task
trips_june = trips_june[['geometry']]
```

```python
trips_june.head()
```

```python
#convert dataset of trips to geodataframe (so it can be merged with the geodataframe of dortmund)
geo_trips_june = gpd.GeoDataFrame(trips_june, crs={'init': 'epsg:4326'}, geometry=trips_june.geometry)
```

```python
# join the data
# merges data when POINT of trips is within POLYGON of a dortmund district
trips_with_postalcode = gpd.sjoin(geo_trips_june, districts_dortmund, how='left', op='within')
```

```python
trips_with_postalcode
```

```python
trips_with_postalcode = trips_with_postalcode.drop(columns = ["index_right", "note", "qkm", "einwohner"])
```

```python
trips_with_postalcode.head()
```

```python
trips_with_postalcode.info()
```

```python
amount_per_postalcode = trips_with_postalcode.groupby("plz").count()
```

```python
amount_per_postalcode
```

```python
amount_per_postalcode = amount_per_postalcode.rename(columns={"geometry": "rental_amount"})
```

```python
# transfer amount of data to the geojson of the district
district_with_count = districts_dortmund.merge(amount_per_postalcode, on='plz', how='left', indicator=True)
```

```python
district_with_count
```

```python
# left_only/NaN => no rentals were done in these districts / there are no stations
```

```python
district_with_count["rental_amount"] = district_with_count["rental_amount"].fillna(0)
```

```python
districts_map = folium.Map(location = [51.5, 7.5], zoom_start=11)

folium.Choropleth(
    geo_data=district_with_count).add_to(districts_map)

# popups instead of tooltips
#for index, row in district_with_count.iterrows():
 #   center = row["geometry"].centroid
  #  center_list = [center.y, center.x]
   # information = row["rental_amount"]
    #folium.Marker(location=center_list, popup=information).add_to(districts_map)
    
folium.GeoJson(
    district_with_count,
    tooltip=folium.features.GeoJsonTooltip(fields=['rental_amount', 'plz'], aliases=['Amount of rentals:', 'Postcal code'])
    
).add_to(districts_map)

display(districts_map)
```
